```python
import pandas as pd 
import matplotlib.pyplot as plt 
import numpy as np 
%matplotlib inline 
import seaborn as sns


#display all the columns of the dataframe 
pd.pandas.set_option('display.max_columns', None)
```

```python
df = pd.read_csv("players.csv")
```

```python
df.head()
```

```python
df.Age.plot(kind="kde")
plt.title("Age Distribution")
plt.xlabel("Age")
plt.show()
```

Most of the players are age between 20 to 30

```python
df.loc[df["Matches"] == df["Matches"].max()]
```

Sofiane Diop from Monaco have played most matches. 

```python
df = df.rename(columns={'Own Goals' : "Own_Goals"})
```

```python
df.loc[df['Own_Goals'] == df["Own_Goals"].max()]
```

```python
OG = df.loc[df['Own_Goals'] == df["Own_Goals"].max()]
OG["Position"].value_counts().plot(kind="barh")
plt.title("Own Goal scored from each position") 
plt.xlabel("Own Goals")
plt.ylabel("Position") 
plt.show()
```

We can see that defenders (LB, RB, CB) score the most own goals

```python
df.loc[df["Red Cards"] == df["Red Cards"].max()] 
```

```python
df = df.rename(columns={'Markey Value In Millions(£)' : 'Market Value In Millions(£)'})
```

```python
df['Position'] = df['Position'].replace({'Centre-Forward': 'CF', 'Left Winger': 'LW', 'Right Winger':'RW',
       'Attacking Midfield' : 'AM', 'Central Midfield' : 'CM', 'Defensive Midfield' : 'DM',
       'Right-Back' : 'RB', 'Centre-Back' : 'CB', 'Second Striker' : 'SS', 'Left-Back' : 'LB',
       'Goalkeeper' : 'GK', 'Left Midfield':'LM', 'Right Midfield' : 'RM' })
```

```python
df['Position'].unique()
```

```python
fig = plt.gcf()

fig.set_size_inches(12, 8)
sns.set(style="ticks", context="talk")
plt.style.use("dark_background")
sns.scatterplot(data=df, x ="Position", y = "Market Value In Millions(£)", hue= "Position")
plt.rcParams.update({"grid.linewidth":0.5, "grid.alpha":0.5})
plt.show()
```

```python
fig = plt.gcf()

fig.set_size_inches(17, 10)
df.groupby('Country')['Goals'].median().plot.bar()
plt.show()
```

Players from these country scored highest average goal

```python
fig = plt.gcf()

fig.set_size_inches(17, 10)
df.groupby('Country')['Assists'].median().plot.bar()
plt.show()
```

Player from these countries have highest average number of assists.

```python
fig = plt.gcf()

fig.set_size_inches(17, 17)
df["Country"].value_counts().plot(kind="barh") 
plt.show()
```

Count of countries players are from.

```python
fig = plt.gcf()

fig.set_size_inches(16, 8)
sns.set(style="ticks", context="talk")
plt.style.use("dark_background")
sns.barplot(data = df, x ="Matches", y = "Goals")

plt.show()
```

Relationship between number of matches and number of goals scored. As we can see as the number of matches increases number of goals scored increases as well.
